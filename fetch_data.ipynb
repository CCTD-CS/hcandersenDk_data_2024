```python
for ref in filtered_items:
    title_element = ref.find('{http://www.tei-c.org/ns/1.0}idno')
    if title_element is not None:
        print(title_element.text)  # This prints the text content of the title element
    else:
        print("@@@@@@@@@@@@@Title not found for item")
```

```python
# Initialize an empty list to store idno text values
idno_texts = []

# Iterate over each item in the filtered_items list
for item in filtered_items:
    # Find the 'idno' element within the item
    idno_element = item.find('{http://www.tei-c.org/ns/1.0}idno')
    
    # Check if the idno element is found
    if idno_element is not None:
        # Extract the text of the idno element and append it to the list
        idno_texts.append(idno_element.text)
    else: print("@@@@@@@@@@@@@Idno not found for item")

print(idno_texts)
```

```python
import os
import xml.etree.ElementTree as ET
all_sentences = []

for tale in idno_texts:
    file_path = f"data/works/{tale}.xml"
    
    # Check if the file exists
    if os.path.exists(file_path):
        tale_file = ET.parse(file_path)
        tale_root = tale_file.getroot()
        text_el = None
        text_tag = "{http://www.tei-c.org/ns/1.0}text"
        for descendant in tale_root.iter(text_tag):
            #print(descendant.tag)
            text_el = descendant

        if text_el is not None:
            for p in text_el.iter("{http://www.tei-c.org/ns/1.0}p"):
                # Remove inner tags from p element
                for child in list(p):  # Use list to create a static copy of children
                    # Append child's text to p's text, if it exists
                    if child.text:
                        p.text = (p.text or '') + child.text
                    # Append child's tail text to p's text, if it exists
                    if child.tail:
                        p.text = (p.text or '') + child.tail
                    # Remove the child element
                    p.remove(child)
                if p.text is not None:
                    all_sentences.append(p.text)
            #print("####################")
    else:
        print(f"File does not exist: {file_path}")
```

```python
all_sentences[:10]
```

```python
# Open a file in write mode
with open('cleaned_data/all_tales_paragrpahs.txt', 'w') as file:
    # Iterate through the list
    for sentence in all_sentences:
        # Write each item followed by a newline
        file.write(sentence + '\n')
```

```python
# remove inner tags in the p tags

# Sample XML data with various child tags within 'p' tags
xml_data = """
<root>
    <p>This is a paragraph with <hi>highlighted text</hi>, <pb>page break</pb>, and <custom>custom tag</custom>.</p>
    <p>Another paragraph with <a>link</a> and <b>bold text</b>.</p>
    <p rend="firstIndent" xml:id="p_00022">»Hvad saa!« svarede Manden, »ja, naar vi frygtede det, saa <seg target="txtcmnt-057-06"><!--DATA-TERM: kneb … i Halsen--><!--DATA-DEFINITION: kværkede, strangulerede.-->kneb vi Dig nu strax i Halsen</seg>, saa Øinene skulde staae ud af Hovedet paa Dig som paa en kogt Krebs! men det frygte vi ikke for; thi til Gavn kan det aldrig blive Dig, og en Ulykke vilde Du være vis paa!«</p>
</root>
"""

# Parse the XML
root1 = ET.fromstring(xml_data)

# Find all 'p' tags
p_tags = root1.findall('.//p')

for p in p_tags:
    for child in list(p):  # Use list to create a static copy of children
        # Append child's text to p's text, if it exists
        if child.text:
            p.text = (p.text or '') + child.text
        # Append child's tail text to p's text, if it exists
        if child.tail:
            p.text = (p.text or '') + child.tail
        # Remove the child element
        p.remove(child)

# Convert back to string to check the result
new_xml = ET.tostring(root1, encoding='unicode')
print(new_xml)
```
