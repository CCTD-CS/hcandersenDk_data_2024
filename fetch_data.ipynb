```python
import os
import xml.etree.ElementTree as ET
from spacy.lang.da import Danish

cleaned_tales = []

sentencizer = Danish()
sentencizer.add_pipe("sentencizer")

# Anchor tags inside seg tags is a challenbge so I use this function to move them.
# This funvtion is generated with copilot
def remove_tei_anchors_preserving_text(xml_root):
    """
    Attempt to remove all TEI 'anchor' elements from the given XML tree while trying to preserve text.
    This function aims to address the issue with text nodes directly following removed elements.

    Parameters:
    - xml_root: The root element of the XML tree.
    """
    prev = None  # Initialize a variable to keep track of the previous element
    for child in list(xml_root):
        if child.tag == '{http://www.tei-c.org/ns/1.0}anchor':
            # Instead of removing the anchor, replace it with its tail text if any
            if child.tail:
                if prev is not None:
                    # Append the anchor's tail text to the previous element's tail
                    prev.tail = (prev.tail or '') + child.tail
                else:
                    # If the anchor is the first child, prepend its tail to the parent's text
                    xml_root.text = (xml_root.text or '') + child.tail
            xml_root.remove(child)
        else:
            remove_tei_anchors_preserving_text(child)
        prev = child  # Update the previous element

for tale in idno_texts:
    file_path = f"data/works/{tale}.xml"
    tale_data = {"title": tale, "text": []}
    # Check if the file exists
    if os.path.exists(file_path):
        tale_file = ET.parse(file_path)
        tale_root = tale_file.getroot()
        text_el = None
        text_tag = "{http://www.tei-c.org/ns/1.0}text"
        for descendant in tale_root.iter(text_tag):
            #print(descendant.tag)
            text_el = descendant

        if text_el is not None:
            remove_tei_anchors_preserving_text(text_el)
            for p in text_el.iter("{http://www.tei-c.org/ns/1.0}p"):
                # Remove inner tags from p element
                for child in list(p):  # Use list to create a static copy of children
                    # Append child's text to p's text, if it exists
                    if child.text:
                        p.text = (p.text or '') + child.text
                    # Append child's tail text to p's text, if it exists
                    if child.tail:
                        p.text = (p.text or '') + child.tail
                    # Remove the child element
                    p.remove(child)
                if p.text is not None:
                    # Split text into sentences with SpaCy sentencizer
                    sen_text = sentencizer(p.text)
                    # Append each sentence to the tale data
                    for sent in sen_text.sents:
                        if sent.text != "\n":
                            tale_data["text"].append(sent.text.strip())
                        else:
                            print("skipped line break inside p element")
        else:
            print(f"No text element in {tale}")
        
        cleaned_tales.append(tale_data)
    else:
        print(f"File does not exist: {file_path}")
```

```python
# remove inner tags in the p tags

from lxml import etree

# Sample XML data with various child tags within 'p' tags
xml_data = """
<root>
<p>1 <seg ><!--DATA-TERM: Oxedriver--><anchor target="txtcmnt-029-01"/><!--primula.-->2</seg> 3</p>
<p>4 <hi>5</hi> 6</p>
</root>
"""

def get_text_excluding_comments(element):
    # Initialize an empty string to hold the combined text
    text = ''
    # Check if the element itself has text and add it
    if element.text:
        text += element.text
    # Iterate through the element's children
    for child in element:
        # If the child is not a comment, process it
        if not isinstance(child, etree._Comment):
            # Recursively get text from child elements, excluding comments
            text += get_text_excluding_comments(child)
        # Add the tail text of the child, if present
        if child.tail:
            text += child.tail
    return text

# Parse the XML
root1 = etree.fromstring(xml_data)

# Iterate through all 'p' elements
for p in root1.findall('.//p'):
    # Use the custom function to get text, excluding comments
    combined_text = get_text_excluding_comments(p)
    # Clear the element's children
    for child in list(p):
        p.remove(child)
    # Update 'p' text with the combined text
    p.text = combined_text.strip()

# Convert back to string to check the result
new_xml = etree.tostring(root1, pretty_print=True, encoding='unicode')
print(new_xml)
```

```python
import xml.etree.ElementTree as ET

def remove_tei_anchors_preserving_text(xml_root):
    """
    Attempt to remove all TEI 'anchor' elements from the given XML tree while trying to preserve text.
    This function aims to address the issue with text nodes directly following removed elements.

    Parameters:
    - xml_root: The root element of the XML tree.
    """
    prev = None  # Initialize a variable to keep track of the previous element
    for child in list(xml_root):
        if child.tag == '{http://www.tei-c.org/ns/1.0}anchor':
            # Instead of removing the anchor, replace it with its tail text if any
            if child.tail:
                if prev is not None:
                    # Append the anchor's tail text to the previous element's tail
                    prev.tail = (prev.tail or '') + child.tail
                else:
                    # If the anchor is the first child, prepend its tail to the parent's text
                    xml_root.text = (xml_root.text or '') + child.tail
            xml_root.remove(child)
        else:
            remove_tei_anchors_preserving_text(child)
        prev = child  # Update the previous element

# The rest of the code remains the same

# Example usage
xml_string = """
<TEI xmlns="http://www.tei-c.org/ns/1.0">
    <text>
        <body>
            <p>This is a paragraph with an <anchor target="cc" xml:id="mark1"/>anchor point.</p>
            <p>Another paragraph with another <anchor xml:id="mark2"/>point.</p>
        </body>
    </text>
</TEI>
"""

# Parse the XML string into an ElementTree, considering the namespace
root = ET.fromstring(xml_string)

# Define the namespace to avoid using it in the tag search explicitly
namespaces = {'tei': 'http://www.tei-c.org/ns/1.0'}
ET.register_namespace('', namespaces['tei'])

# Remove all TEI 'anchor' elements
remove_tei_anchors_preserving_text(root)

# Convert the modified tree back to a string and print it
print(ET.tostring(root, encoding='unicode'))
```
