```python
# Initialize an empty list to store idno text values
idno_texts = []

# Iterate over each item in the filtered_items list
for item in filtered_items:
    # Find the 'idno' element within the item
    idno_element = item.find('{http://www.tei-c.org/ns/1.0}idno')
    
    # Check if the idno element is found
    if idno_element is not None:
        # Extract the text of the idno element and append it to the list
        idno_texts.append(idno_element.text)
    else: print("@@@@@@@@@@@@@Idno not found for item")

print(idno_texts[:5])
```

```python
import os
import xml.etree.ElementTree as ET
from spacy.lang.da import Danish

cleaned_tales = []

sentencizer = Danish()
sentencizer.add_pipe("sentencizer")

# Anchor tags inside seg tags is a challenbge so I use this function to move them.
# This funvtion is generated with copilot
def remove_tei_anchors_preserving_text(xml_root):
    """
    Attempt to remove all TEI 'anchor' elements from the given XML tree while trying to preserve text.
    This function aims to address the issue with text nodes directly following removed elements.

    Parameters:
    - xml_root: The root element of the XML tree.
    """
    prev = None  # Initialize a variable to keep track of the previous element
    for child in list(xml_root):
        if child.tag == '{http://www.tei-c.org/ns/1.0}anchor':
            # Instead of removing the anchor, replace it with its tail text if any
            if child.tail:
                if prev is not None:
                    # Append the anchor's tail text to the previous element's tail
                    prev.tail = (prev.tail or '') + child.tail
                else:
                    # If the anchor is the first child, prepend its tail to the parent's text
                    xml_root.text = (xml_root.text or '') + child.tail
            xml_root.remove(child)
        else:
            remove_tei_anchors_preserving_text(child)
        prev = child  # Update the previous element

for tale in idno_texts:
    file_path = f"data/works/{tale}.xml"
    tale_data = {"title": tale, "text": []}
    # Check if the file exists
    if os.path.exists(file_path):
        tale_file = ET.parse(file_path)
        tale_root = tale_file.getroot()
        text_el = None
        text_tag = "{http://www.tei-c.org/ns/1.0}text"
        for descendant in tale_root.iter(text_tag):
            #print(descendant.tag)
            text_el = descendant

        if text_el is not None:
            remove_tei_anchors_preserving_text(text_el)
            for p in text_el.iter("{http://www.tei-c.org/ns/1.0}p"):
                # Remove inner tags from p element
                for child in list(p):  # Use list to create a static copy of children
                    # Append child's text to p's text, if it exists
                    if child.text:
                        p.text = (p.text or '') + child.text
                    # Append child's tail text to p's text, if it exists
                    if child.tail:
                        p.text = (p.text or '') + child.tail
                    # Remove the child element
                    p.remove(child)
                if p.text is not None:
                    # Split text into sentences with SpaCy sentencizer
                    sen_text = sentencizer(p.text)
                    # Append each sentence to the tale data
                    for sent in sen_text.sents:
                        if sent.text != "\n":
                            tale_data["text"].append(sent.text.strip())
                        else:
                            print("skipped line break inside p element")
        else:
            print(f"No text element in {tale}")
        
        cleaned_tales.append(tale_data)
    else:
        print(f"File does not exist: {file_path}")
```

```python
# Open a file in write mode
for tale in cleaned_tales:
    with open(f'cleaned_data/{tale["title"]}.txt', 'w') as file:
        # Iterate through the list
        for sentence in tale["text"]:
            # Write each item followed by a newline
            file.write(sentence + '\n')
```
